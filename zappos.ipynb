```python


brand_name = []
item_name = []
price = []
old_price = []
ratings = []
review = []
link = []

sleep(5)
pagination = driver.find_element_by_xpath('//*[@id="searchPagination"]')
pages = pagination.find_elements_by_xpath('.//a[contains(@title, "Page")]')

last_page = int(pages[-1].text)
current_page = 1
while current_page <= last_page:
    shoes = WebDriverWait(driver, 50).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="products"]/article')))
    # shoes = driver.find_elements_by_xpath('//*[@id="products"]/article')
    print("Srapping data...\n")
    
    for shoe in shoes:
        try:
            brand_name.append(shoe.find_element_by_xpath('.//dd[@itemprop="brand"]').text)
        except:
            brand_name.append(np.nan)
        try:
            item_name.append(shoe.find_element_by_xpath('.//dd[@itemprop="name"]').text)
        except:
            item_name.append(np.nan)
        try:
            price.append(shoe.find_element_by_xpath('.//dd[@itemprop="offers"]/span[@itemprop="price"]').text)
        except:
            price.append(np.nan)
        try:
            old_price.append(shoe.find_element_by_xpath('.//dd[@itemprop="offers"]/span[2]/span[2]').text)
        except:
            old_price.append(np.nan)
        try:
            ratings.append(shoe.find_element_by_xpath('.//dd/span[@itemprop="aggregateRating"]').text)
        except:
            ratings.append(np.nan)
        try:
            review.append(shoe.find_element_by_xpath('.//dd/span[contains(@aria-label, "reviews")]').text)
        except:
            review.append(np.nan)
        try:
            link.append(shoe.find_element_by_xpath('./a[@itemprop="url"]').get_attribute('href'))
        except:
            link.append(np.nan)
    current_page = current_page + 1
    
    # try:
        
    next_page = driver.find_element_by_xpath('//*[@id="searchPagination"]//a[@rel="next"]')
        # Wait for the next page to load before proceeding
    # next_page =  WebDriverWait(driver, 30).until(EC.presence_of_element_located(By.XPATH, '//*[@id="searchPagination"]//a[@rel="next"]'))
    #     # WebDriverWait(driver, 10).until(EC.staleness_of(products[0]))
    next_page.click()
    print(f"Page {current_page} has been successfully scraped")
    # except:
    #     pass
    #     print(f"Page {current_page} cannot be scraped")


```

```python
df = pd.DataFrame({
"brand_name" : brand_name,
"item_name" : item_name,
"price" : price,
"old_price" : old_price,
"ratings" : ratings,
"reviews" : review,
"link" : link
})
print('Dataframe as been Created Successfully ... \n')
```
